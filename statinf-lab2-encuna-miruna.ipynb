Q33. Din teorema Bayes, avem:

$$P(D|+) = \frac{P(+|D)P(D)}{P(+|D)P(D)+P(+|D^c)P(D^c)}$$

Deoarece $P(D^c) = 1 - P(D)$ i $P(+|D^c) = 1 - P(-|D^c)$ (elemente complementare), putem calcula probabilitatea ca chiar s aib boala dac testul iese pozitiv, cunoscnd senzitivitatea (rezultat pozitiv cu boal) de 99.7% i specificitatea (test negativ fr boal) de 98.5%.

Probabilitatea ca pacientul s aib boala dac testul e pozitiv, $P(D|+)$ (positive predictive value) este:

```python
# rotunjii la 4 zecimale folosind funcia round() 
# (apsai Shift-Tab cu cursorul dup paranteza deschis la 'round(' ca s vedei cum se folosete)
def q():
    sensitivity = 0.997   # senzitivitatea este 99.7%, rezultat pozitiv cu boala
    specificity = 0.985   # specificitatea este 98.5%  test negativ fara boala
    prevalence = 0.001    # prevalena bolii n populaie este 0,1%
    
    #probabilitatea c un pacient are boala
    p_disease = prevalence
    
    #probabilitatea c un pacient nu are boala
    p_no_disease = 1 - p_disease
    
    #probabilitatea unui rezultat pozitiv dac pacientul are boala
    p_pos_disease = sensitivity
    
    #probabilitatea unui rezultat pozitiv dac pacientul nu are boala
    p_pos_no_disease = 1 - specificity
    
    #probabilitatea c pacientul are boala dac testul este pozitiv
    p_disease_given_pos = (p_pos_disease * p_disease) / ((p_pos_disease * p_disease) + (p_pos_no_disease * p_no_disease))
    
    return round(p_disease_given_pos, 4)

print('Raspuns %s' % ('corect' if r(33, q()) else 'incorect'))
```
